```python
import numpy as np

test = [
    [1,2,3,4],  # Index 0
    [5,6,7,8],  # Index 1
    [7,8,9,5],  # Index 2
]

# 첫 번째 리스트(인덱스 0)의 인덱스 1 값을 4로 변경
test[0][1] = 4

# 두 번째 리스트(인덱스 1)의 인덱스 1 값을 4로 변경
test[1][1] = 4

# 출력: [[1, 4, 3, 4], [1, 4, 3, 4]]

np_test = np.array(test)
mask = np_test[:,1] > 5
print(np_test[:,1] )
print(mask)  # 출력: [False False  True]
print(np_test[mask])  # 출력: [6 8]
```
